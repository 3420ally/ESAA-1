### 7. 각 나라의 에볼라 발병일을 같은 시점으로 옮겨 비교하기

```python
#각 column별로 마지막 valid index 구하기 (마지막 -> 제일 최초 발병일)
last_valid = ebola.apply(pd.Series.last_valid_index)
```

```python
#각 나라별로 첫 valid index date 구하기
first_valid = ebola.apply(pd.Series.first_valid_index)
```

```python
last_valid #각 나라별 최초 발병일
```

```python
first_valid ##각 나라별 마지막 발병일
```

```python
earliest_date = ebola.index.min() 
print(earliest_date)
```

```python
#전세계적 최초 발병 날짜로부터 나라별 최초 발병날짜까지의 일수 구하기
shift_values = last_valid - earliest_date
print(shift_values)
```

```python
#각 나라의 최초 발병일을 동일하게 한 dataframe다시 만들기
ebola_dict = {}
for idx, col in enumerate(ebola): #enumerate가 ebola의 컬럼값을 반환할건데 그 값을 인덱스와 같이 반환하겠다
    d = shift_values[idx].days
    shifted = ebola[col].shift(d) #d만큼 shift 해줘라
    ebola_dict[col] = shifted
```

```python
ebola_shift = pd.DataFrame(ebola_dict) #데이터 프레임화 해주기
```

```python
ebola_shift.tail()
```

```python
ebola_shift.index = ebola_shift['Day'] 
ebola_shift = ebola_shift.drop(['Date', 'Day'], axis=1)

ebola_shift.tail()
```

```python
fig, ax = plt.subplots(figsize=(8,6)) 
ax = ebola_shift.iloc[:, :].plot(ax=ax)
ax.legend(fontsize=7, loc=2, borderaxespad=0.) 

#plt.ylim(0,2000) #잘안보이니 이것을 적용한 후 시각화
plt.show()
```
